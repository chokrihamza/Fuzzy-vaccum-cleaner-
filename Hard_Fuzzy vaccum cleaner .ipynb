```python
import numpy as np 
import skfuzzy as fuzz 
import matplotlib.pyplot as plt 
```

# Antecedents and Consequent

```python
x_surface = np.arange( 0 , 6 , 1)
x_dirt = np.arange(0 ,6 ,1)
x_suction = np.arange(0 , 11 ,1)
```

```python
x_surface,x_dirt,x_suction
```

```python
# Suface 
```

```python
y_surface_Easy = fuzz.trimf(x_surface , [0 ,0 ,2.5])
y_surface_Moderate  = fuzz.trimf(x_surface , [0 ,2.5 ,5])
y_surface_Difficult = fuzz.trimf(x_surface , [2.5 ,5 ,5])
```

```python
y_surface_Easy ,y_surface_Moderate ,y_surface_Difficult
```

```python
x_surface ,y_surface_Easy
```

```python
plt.plot(x_surface, y_surface_Easy ,'ro', label ='easy')
plt.plot(x_surface, y_surface_Easy)
plt.legend()
```

```python
x_surface ,y_surface_Moderate
```

```python
plt.plot(x_surface, y_surface_Moderate ,'go', label ='Moderate')
plt.plot(x_surface, y_surface_Moderate)
plt.legend()
```

```python
x_surface ,y_surface_Difficult
```

```python
plt.plot(x_surface, y_surface_Difficult ,'bo', label ='Difficult')
plt.plot(x_surface, y_surface_Difficult)
plt.legend()
```

```python
fig ,ax0 = plt.subplots(figsize= (8 ,5 ))
ax0.plot(x_surface, y_surface_Easy ,'b', label ='Easy')
ax0.plot(x_surface, y_surface_Moderate ,'g', label ='Moderate')
ax0.plot(x_surface, y_surface_Difficult ,'r', label ='Difficult')
ax0.set_title('Surface')
ax0.legend()
```

# Dirt 

```python
y_dirt_Light = fuzz.trimf(x_dirt , [0 ,0 ,2.5])
y_dirt_Moderate  = fuzz.trimf(x_dirt , [0 ,2.5 ,5])
y_dirt_Heavy = fuzz.trimf(x_dirt , [2.5 ,5 ,5])
```

```python
y_dirt_Light,y_dirt_Moderate,y_dirt_Heavy
```

```python
y_suction_Low = fuzz.trimf(x_suction , [0 ,0 ,5])
y_suction_Medium  = fuzz.trimf(x_suction , [0 , 5 ,10])
y_suction_Heigh= fuzz.trimf(x_suction , [5 ,10,10])
```

```python
y_suction_Low ,y_suction_Medium , y_suction_Heigh
```

```python
fig ,ax2 = plt.subplots(figsize= (8 ,5 ))
ax2.plot(x_suction, y_suction_Low ,'b', label ='Low')
ax2.plot(x_suction, y_suction_Medium ,'g', label ='Medium')
ax2.plot(x_suction, y_suction_Heigh ,'r', label ='Heigh')
ax2.set_title('Suction')
ax2.legend();
```

# inputs

```python
surface_level_easy = fuzz.interp_membership(x_surface,y_surface_Easy, 4.0)
surface_level_moderate = fuzz.interp_membership(x_surface,y_surface_Moderate, 4.0)
surface_level_difficult = fuzz.interp_membership(x_surface,y_surface_Difficult, 4.0)
```

```python
surface_level_easy ,surface_level_moderate ,surface_level_difficult 
```

```python
dirt_level_light = fuzz.interp_membership(x_dirt, y_dirt_Light, 3.0)
dirt_level_moderate = fuzz.interp_membership(x_dirt, y_dirt_Moderate, 3.0)
dirt_level_heavy = fuzz.interp_membership(x_dirt, y_dirt_Heavy, 3.0)
```

```python
dirt_level_light,dirt_level_moderate,dirt_level_heavy
```

# rules 

## rule 1

####      If the surface is smooth (easy ) and the dirt is light then the suction will be low.

```python
surface_level_easy ,dirt_level_light
```

```python
rule1 = np.fmax(surface_level_easy , dirt_level_light)
```

```python
rule1
```

```python
activate_suction_low = np.fmin(rule1 , y_suction_Low )
```

```python
y_suction_Low
```

```python
rule1 ,activate_suction_low
```

## graph with intersections

```python
x_suction0 = np.zeros_like(x_suction)
```

```python
x_suction0
```

```python
fig ,ax3 = plt.subplots(figsize= (8 ,3 ))
ax3.fill_between (x_suction,x_suction0,activate_suction_low ,facecolor= 'b', alpha = 0.7)
ax3.plot(x_suction, y_suction_Low ,'b',linewidth =0.5 ,linestyle='--', label ='Low')
ax3.fill_between (x_suction,x_suction0,activate_suction_medium ,facecolor= 'g', alpha = 0.7)
ax3.plot(x_suction, y_suction_Medium ,'g', label ='Medium',linewidth =0.5 ,linestyle='--')
ax3.fill_between (x_suction,x_suction0,activate_suction_heigh ,facecolor= 'r', alpha = 0.7)
ax3.plot(x_suction, y_suction_Heigh ,'r', label ='Heigh',linewidth =0.5 ,linestyle='--')
ax3.set_title('Departure from interconnected activities')
ax3.legend();
```
